```python
import torch
import numpy as np
```

```python
x = torch.empty(1)   #tensor of shape 1
y = torch.empty(5)   #tensor of shape 5
print("x is {}".format(x))
print("y is {}".format(y))
```

```python
x = torch.empty(2,3)   #tensor of shape (2,3)
print("x is {}".format(x))
print("Dtype of x is {}".format(x.dtype))  #by default float32

print()

y = torch.empty(2,3,2)   #tensor of shape (2,3,2)
print("y is {}".format(y))
```

```python
x = torch.rand(2,2)   #random
print("x is {}".format(x))

print()

y = torch.zeros(3,3)    #zeros
print("y is {}".format(y))

print()

z = torch.ones(3,3)   #ones
print("z is {}".format(z))

print()

z2 = torch.ones(3,3)*2   #multiplying constant with tensor
print("z*2 is {}".format(z2))

print()

z3 = torch.ones(3,3) + 3   #adding constant to tensor
print("z+3 is {}".format(z3))
```

```python
z = torch.ones(3,3, dtype=torch.double) #ones but in double dtype
print("z is {}".format(z))

z = torch.ones(3,3, dtype=torch.float32) #ones but in float32 dtype (default)
print("z is {}".format(z))

z = torch.ones(3,3, dtype=torch.float16) #ones but in float16 dtype
print("z is {}".format(z))

z = torch.ones(3,3, dtype=torch.int) #ones but in int dtype
print("z is {}".format(z))
```

```python
x= torch.tensor([1,3,4,7])      #list to tensor
y = torch.tensor(np.array([[100,200],
                           [300,400]]))  #numpy array to tensor

print("x is {}".format(x))
print()
print("y is {}".format(y))
```

```python
x= torch.ones(3,3)*2
y= torch.ones(3,3)+5
print("x is {}".format(x))
print("y is {}".format(y))

z1 = x + y     #elementwise addition
print("z1 is {}".format(z1))

z2 = x * y   #elementwise multiplication
print("z2 is {}".format(z2))

z3 = x / y   #elementwise divisiion
print("z3 is {}".format(z3))
```

```python
x= torch.ones(3,3)*2
y= torch.ones(3,3)+5
print("x is {}".format(x))
print("y is {}".format(y))

print("Addition= {}".format(torch.add(x,y)))
print("Subtraction= {}".format(torch.sub(x,y)))   #sub or subtract (x-y)
print("Multiplication= {}".format(torch.mul(x,y)))  #mul or multiply (x*y)
print("Division= {}".format(torch.div(x,y)))   #div or divide (x/y)
```

```python
x= torch.ones(3,3)*2
y= torch.ones(3,3)+5
print("x is {}".format(x))
print("y is {}".format(y))

y.add_(x)  #same as y=y+x
print("Modified (added) y = {}".format(y))

y.sub_(x)   #same as y=y-x
print("Modified (subtracted) y = {}".format(y))

y.multiply_(x)    #same as y=y*x
print("Modified (multiplied) y = {}".format(y))

y.divide_(x)   #same as y=y/x
print("Modified (divided) y = {}".format(y))


#In pytorch, every function with a trailing underscore means
#in place operation
```

```python
x= torch.ones(3,3)*2
y= torch.ones(3,3)+5
print("x is {}".format(x))
print("y is {}".format(y))

y+= x   #similiarly -,*,/
print(y)
```

```python
x = torch.rand(4,6)
print(x)
print("All rows of 3rd column: {}".format(x[:,3]))    #slicing like in numpy
print("All rows of 1st two columns: {}".format(x[:,:2]))
print("All rows of last 3 columns: {}".format(x[:,-3:]))
```

```python
x = torch.rand(3,4)
print(x)
print(x[2,2])
print(x[2,2].item()) #item()off used only if 1 value is to be needed
```

```python
x= torch.rand(4,3)
print(x)
print(x.view(6,2))  #reshaped into 6,2
print(x.view(12))   #reshaped into 12
```

```python
x= torch.rand(4,3)
print(x)
y= x.view(-1,2)   #when one or more dimension is unknown
print(y)
print(y.size())
```

```python
x = torch.rand(3,5)
y = x.numpy()   #tensor to numpy
print(y)
print(type(y))
```

```python
x = torch.zeros(3,5)
y = x.numpy()   #torch to numpy

x.add_(1)  #y is also added to 1
print(y)
```

```python
y = np.zeros((3,3))
x = torch.from_numpy(y)  #numpy to torch
print(x)

y+=1     #x also gets added to one
print(x)
```

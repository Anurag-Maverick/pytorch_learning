```python
import torch
import numpy as np
```

```python
a = torch.randn(5, requires_grad=True)
ab = ((a + 2) ** 2).sum()
ab.backward()
print(a.grad)
```

```python
x = torch.tensor([.13,0.015,0.17], requires_grad=True, dtype=torch.float32)
y = x*6
z = y*y/2
z = z.mean()
print(z)

z.backward()
print(z)
print(x.grad)
```

Stopping pytorch from tracking gradient

```python
x = torch.tensor([.13,0.015,0.17], requires_grad=True, dtype=torch.float32)

x.requires_grad_(False)  #disables grading tracking
y = x**2

print(x)
print(y)
```

```python
x = torch.tensor([.13,0.015,0.17], requires_grad=True, dtype=torch.float32)
y = x.detach()
print(y)
```

```python
x = torch.tensor([.13,0.015,0.17], requires_grad=True, dtype=torch.float32)
with torch.no_grad():
    y = x + 2
    print(y)
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(3):
    model_output = (weights*5).sum()
    model_output.backward() #accumulates the previous values in grad attribute as many times as called
    print(weights.grad)
    
    #accumulations results in incorrect gradient with each epoch
```

```python
weights = torch.ones(4, requires_grad=True)
for epoch in range(3):
    model_output = (weights*5).sum()
    model_output.backward() 
    print(weights.grad)
    weights.grad.zero_()  #empties gradient, hence avoiding acculumation
    
#doesnot accumulate here, so sorrect!
```
